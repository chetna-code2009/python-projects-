```python
def great():

    x = int(input("Enter number 1: "))
    y = int(input("Enter number 2: "))
    z = int(input("Enter number 3: "))

    print("Greatest number is:", max(x, y, z))


great()
```

```python
import random

def play_coin_toss():
    print("--- Welcome to the Ultimate Coin Toss Game! --")
    player_score = 0
    computer_score = 0
     # Main game loop
    while True:
        play_round = input("\nDo you want to toss the coin? (y/n): ").strip().lower()

        if play_round != 'y':
            print("\nThanks for playing!")
            print(f"Final Score -> You: {player_score} | Computer: {computer_score}")
            break
        guess = input("Guess the outcome (head/tail): ").strip().lower()
        if guess not in ["head", "tail"]:
            print("Invalid input! Please type 'head' or 'tail'.")
            continue
        result = random.choice(["head", "tail"])
        print(f"The coin landed on: {result.upper()}")
        if guess == result:
            print("🎉 You guessed right! You get a point.")
            player_score += 1
        else:
            print("❌ Wrong guess! The computer gets a point.")
            computer_score += 1
        print(f"Current Score -> You: {player_score} | Computer: {computer_score}")
play_coin_toss()
```
