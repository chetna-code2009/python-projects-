```python
import random

def play_coin_toss():
    print("--- Welcome to the Ultimate Coin Toss Game! --")
    player_score = 0
    computer_score = 0
     # Main game loop
    while True:
        play_round = input("\nDo you want to toss the coin? (y/n): ").strip().lower()

        if play_round != 'y':
            print("\nThanks for playing!")
            print(f"Final Score -> You: {player_score} | Computer: {computer_score}")
            break
        guess = input("Guess the outcome (head/tail): ").strip().lower()
        if guess not in ["head", "tail"]:
            print("Invalid input! Please type 'head' or 'tail'.")
            continue
        result = random.choice(["head", "tail"])
        print(f"The coin landed on: {result.upper()}")
        if guess == result:
            print("🎉 You guessed right! You get a point.")
            player_score += 1
        else:
            print("❌ Wrong guess! The computer gets a point.")
            computer_score += 1
        print(f"Current Score -> You: {player_score} | Computer: {computer_score}")
play_coin_toss()
```

```python
def calculate_bciit_semester_percentage():
    print("=== BCIIT BCA 1st Year Advanced Performance Evaluator ===")

    bca_s1_subjects = {
        "BCA 101 Discrete Mathematics": 4,
        "BCA 103 Programming using C": 4,
        "BCA 105 Fundamentals of IT": 3,
        "BCA 107 Web Technologies": 3,
        "BCA 151 C Programming Lab": 2,
        "BCA 153 IT Lab": 2
    }

    total_credits = sum(bca_s1_subjects.values())
    earned_total_marks = 0
    max_total_marks = len(bca_s1_subjects) * 100
    subject_reports = {}

    print(f"\nEnter marks for {len(bca_s1_subjects)} Semester-1 papers:")

    for subject, credits in bca_s1_subjects.items():
        print(f"\n--- {subject} ---")

        # Defensive Input Validation Loop
        while True:
            try:
                internal = float(input("Enter Internal Marks (Out of 40): "))
                external = float(input("Enter External End-Term Marks (Out of 60): "))

                if 0 <= internal <= 40 and 0 <= external <= 60:
                    break
                print("❌ Invalid Entry! Internals must be 0-40 and Externals 0-60.")
            except ValueError:
                print("❌ Invalid data type! Please input numerical integers or floats.")

        # Aggregate marks out of 100
        final_score = internal + external
        earned_total_marks += final_score

        # Determine GGSIPU Passing/Failing conditions (Minimum 40 aggregate to pass)
        status = "PASSED" if final_score >= 40 else "FAILED (BACKLOG)"

        # Calculate localized point index for future SGPA additions
        if final_score >= 90: grade, points = "O (Outstanding)", 10
        elif final_score >= 80: grade, points = "A+ (Excellent)", 9
        elif final_score >= 70: grade, points = "A (Very Good)", 8
        elif final_score >= 60: grade, points = "B+ (Good)", 7
        elif final_score >= 50: grade, points = "B (Above Average)", 6
        elif final_score >= 40: grade, points = "C (Pass)", 5
        else: grade, points = "F (Fail)", 0

        subject_reports[subject] = {
            "Total": final_score,
            "Grade": grade,
            "Status": status
        }

    # 2. Compute Metric Calculations
    overall_percentage = (earned_total_marks / max_total_marks) * 100

    # 3. Print Comprehensive Performance Analytics Breakdown
    print("\n================ OFFICIAL PERFORMANCE REPORT ================")
    for sub, report in subject_reports.items():
        print(f"{sub:30} | Score: {report['Total']:5.1f}/100 | Grade: {report['Grade']:18} | Status: {report['Status']}")

    print("=============================================================")
    print(f"Total Marks Gathered : {earned_total_marks:.1f} / {max_total_marks}")
    print(f"Final Aggregated Score: {overall_percentage:.2f}%")

    # Highlight performance assessment metric
    if overall_percentage >= 75:
        print("🏆 Academic Standing: Distinction Status")
    elif overall_percentage >= 60:
        print("🏅 Academic Standing: First Division Status")
    elif overall_percentage >= 40:
        print("👍 Academic Standing: Pass Classification")
    else:
        print("🚨 Academic Standing: Unsatisfactory / Critical Retake Required")
    print("=============================================================")

# Run the system implementation block
calculate_bciit_semester_percentage()
```
