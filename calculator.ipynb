```python
previous_answer = None

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero."
    return a / b

def percentage(a, b):
    return (a * b) / 100


operations = {
    "+": add,
    "-": subtract,
    "*": multiply,
    "/": divide,
    "%": percentage
}


while True:
    print("\n--- Python Calculator ---")

    if previous_answer is not None:
        print(f"Previous answer: {previous_answer}")

    try:
        first_input = input(
            "Enter the first number or type 'ans' to use the previous answer: "
        ).strip().lower()

        if first_input == "ans":
            if previous_answer is None:
                print("No previous answer available.")
                continue
            first_number = previous_answer
        else:
            first_number = float(first_input)

        operator = input("Enter operation (+, -, *, /, %): ").strip()

        if operator not in operations:
            print("Invalid operation.")
            continue

        second_number = float(input("Enter the second number: "))

        result = operations[operator](first_number, second_number)

        if isinstance(result, str):
            print(result)
            continue

        previous_answer = result

        print(f"Result: {result}")

    except ValueError:
        print("Error: Please enter a valid number.")

    again = input("\nDo you want to calculate again? (yes/no): ").strip().lower()

    if again != "yes":
        print("Calculator closed.")
        break
```
