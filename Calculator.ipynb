```python
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero."
    return a / b

def percentage(a, b):
    # Calculates b% of a (e.g., 10% of 200 = 20)
    return (a * b) / 100

operations = {
    '+': add,
    '-': subtract,
    '*': multiply,
    '/': divide,
    '%': percentage,
}

def calculate(num1, operator, num2):
    if operator not in operations:
        return None
    function = operations[operator]
    return function(num1, num2)

def get_number_input(prompt):
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Invalid input. Please enter a number.")

# This variable stores the memory of your last answer
previous_answer = None

while True:
    # 1. Decide whether to use the previous answer or a new first number
    if previous_answer is not None:
        print(f"\nPrevious answer saved in memory: {previous_answer}")
        use_past = input("Use this answer as your first number? (yes/no): ").lower()
        if use_past == 'yes' or use_past == 'y':
            a = previous_answer
        else:
            a = get_number_input("Enter first number: ")
    else:
        a = get_number_input("\nEnter first number: ")

    # 2. Get the operation symbol
    print("Available operations: +, -, *, /, %")
    operation_symbol = ''
    while operation_symbol not in operations:
        operation_symbol = input("Pick an operation: ")
        if operation_symbol not in operations:
            print("Invalid operator. Try again.")

    # 3. Get the second number
    if operation_symbol == '%':
        b = get_number_input(f"What percentage of {a} do you want? (Enter a number): ")
    else:
        b = get_number_input("Enter second number: ")

    # 4. Calculate and show result
    result = calculate(a, operation_symbol, b)

    if isinstance(result, str): # Checks if the result is an error message string
        print(f"\n{result}")
        previous_answer = None # Reset memory if an error occurs
    else:
        print(f"\nResult: {a} {operation_symbol} {b} = {result}")
        previous_answer = result # Save the successful answer to memory

    # 5. Check if the user wants to quit entirely
    repeat = input("\nDo you want to use the calculator again? (yes/no): ").lower()
    if repeat != 'yes' and repeat != 'y':
        print("Goodbye!")
        break
```
